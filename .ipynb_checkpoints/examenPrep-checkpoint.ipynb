## PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
cols=[]
for i,typex in enumerate(df.dtypes):
    if(typex!='object'):
        cols.append(df.columns[i])
X=df[cols].values
x = StandardScaler().fit_transform(X)
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
pca.explained_variance_
```

```python
ax=sns.scatterplot(x=principalComponents[:,0],y=principalComponents[:,1])
ax.set_xlabel('Componente principal 1')
ax=ax.set_ylabel('Componente principal 2')
```

```python
df['pca1']=principalComponents[:,0]
df['pca2']=principalComponents[:,1]
corr=df.corr().values
ax=sns.heatmap(corr[:-2,-2:],cmap='RdBu',yticklabels=cols[:-2],xticklabels=['Principal component 1','Principal component 2'],annot=True,center=0)
```

## K-means

```python
sns.set()
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3).fit(principalComponents)
labels=kmeans.predict(principalComponents)
centers=kmeans.cluster_centers_
fig,ax=plt.subplots()
ax=sns.scatterplot(x=principalComponents[:,0],y=principalComponents[:,1],hue=labels+1,zorder=0,legend='full',ax=ax,palette='tab20b')
leg=ax.legend(title='Grupos')
sns.scatterplot(x=centers[:,0],y=centers[:,1],s=500,hue=[0,1,2],legend=None,zorder=10,ax=ax,palette='tab20b')
ax.set_xlabel('Componente principal 1')
ax=ax.set_ylabel('Componente principal 2')
```

## LDA

```python
from sklearn import datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.colors as colors
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits import mplot3d
from sklearn import linear_model, datasets
import seaborn as sns
import itertools

%matplotlib inline
sns.set()
#plt.style.use('fivethirtyeight')

plt.rcParams["figure.figsize"] = (20,10)




iris = datasets.load_iris()
fig,ax=plt.subplots()

X = iris.data
y = iris.target
target_names = iris.target_names

lda = LinearDiscriminantAnalysis(n_components=2)
X_lda= lda.fit(X, y).transform(X)

colorss = ['navy', 'turquoise', 'darkorange']

for color, i, target_name in zip(colorss, [0, 1, 2], target_names):
    ax.scatter(X_lda[y == i, 0], X_lda[y == i, 1], alpha=.7, color=color,
                label=target_name)
ax.legend(loc='best', shadow=False, title='Especies')

#### a ver si jala
lda2 = LinearDiscriminantAnalysis(n_components=2)
X_data = X_lda
y_labels = y
lda2.fit(X_data,y_labels)





# Our 2-dimensional distribution will be over variables X and Y
N = 100
X = np.linspace(-10, 10, N)
Y = np.linspace(-3, 3, N)
X, Y = np.meshgrid(X, Y)


color_list = ['Blues','Greens','Reds']
my_norm = colors.Normalize(vmin=-1.,vmax=1.)



#Computing the predicted class function for each value on the grid
zz = np.array(  [lda2.predict( np.array([[xx,yy]])) 
                     for xx, yy in zip(np.ravel(X), np.ravel(Y)) ] )
    
#Reshaping the predicted class into the meshgrid shape
Z = zz.reshape(X.shape)


#Plot the filled and boundary contours
ax.contourf( X, Y, Z, 2, alpha = .1, colors =colorss)
ax.contour( X, Y, Z, 2, alpha = 1, colors = colorss)
# Addd axis and title
ax.set_xlabel('Función discriminante 1')
ax.set_ylabel('Función discriminante 2')
ax.set_title('LDA and boundaries')

plt.show()
```
